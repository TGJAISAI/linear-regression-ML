```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")
df.head(5)
```

```python
df.rename(columns = {'per capita income (US$)':'PCI'}, inplace = True)
```

```python
df.head()
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.	year,df.PCI,color="red",marker="+")
```

```python
new_df = df.drop("year",axis="columns")
new_df
```

```python
year=df.year
year
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(new_df,year)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
